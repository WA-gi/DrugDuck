```python
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
import seaborn as sns
sns.set(style='ticks')
import matplotlib.pyplot as plt
```

```python
# Target search for MDM2-p53
target = new_client.target
target_query = target.search('Staphylococcus aureus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[0]
selected_target
```

```python

activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
Staph_data = pd.DataFrame.from_dict(res)
Staph_data.info()
```

```python
Staph_data.to_csv('Staph_data_bioactivity_data_raw.csv', index=False)
Staph_data
```

```python
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
Staph_data_s= Staph_data[selection]
Staph_data_s.to_csv('Staph_bioactivity_data.csv', index=False)
Staph_data_s
```

```python
Staph_data2 = Staph_data_s.drop_duplicates(['molecule_chembl_id'])
Staph_data2
```

```python
Staph_data2 = Staph_data2.astype({'standard_value': 'float64'})
Staph_data2.info()
```

```python
bioactivity_threshold = []
for i in Staph_data2.standard_value:
  if float(i) >= 10000:
    bioactivity_threshold.append("inactive")
  elif float(i) <= 1000:
    bioactivity_threshold.append("active")
  else:
    bioactivity_threshold.append("intermediate")
```

```python
bioactivity_class = pd.Series(bioactivity_threshold, name='class')
bioactivity_class
```

```python
mol_id = []
for i in Staph_data2.molecule_chembl_id:
    mol_id.append(i)
canonical_smiles = []
for i in Staph_data2.canonical_smiles:
    canonical_smiles.append(i)   
standard_value = []
for i in Staph_data2.standard_value:
    standard_value.append(i)
    
selection = list(zip(mol_id, canonical_smiles, standard_value, bioactivity_class))
Staph_data3 = pd.DataFrame( selection, columns = ['mol_id', 'canonical_smiles', 'standard_value', 'bioactivity_class'])
Staph_data3.info()
```

```python
Staph_data4 = Staph_data3.dropna()
Staph_data4.info()
```

```python
Staph_data4.to_csv('Staph_data4_bioactivity.csv', index=False)
```
